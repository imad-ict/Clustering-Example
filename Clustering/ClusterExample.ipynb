# LIBRARIES

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
%matplotlib inline
```

# Read CSV

```python
states = pd.read_csv(r'C:\Users\Imad\Desktop\Lynda Data Science\Clustering\ClusterData.csv')
list(states.columns.values)
```

```python
states.head()
```

```python
print(states.shape)
```

## Save numerical data only

```python
st = states[states.columns[2:]]
st.index = states.ix[:,1]
```

# Linkage Matrix

```python
X = linkage(st, 'ward')
```

# Dendrogram of Clusters (Devisive)

```python
plt.figure(figsize = (25, 10))
plt.title('Cluster with All Searches and Personality')
plt.ylabel('distance')
dendrogram(
    X,
    labels = st.index,
    leaf_rotation = 0.5,
    leaf_font_size = 16.,
)
plt.show()
```
